Step 1: import the package

```python
import pandas as pd
import biogeme.database as db
import biogeme.biogeme as bio
```

Step 2: prepare the data

```python
pandas = pd.read_table("AttributesPerceptions.dat")
pandas
```

```python
pandas.describe()
```

```python
database = db.Database("Attributes",pandas)
```

import the headers for future use in the formulas

```python
from headers import *
```

Model specification

Parameters to be estimated

```python
B_LDW = Beta('B_LDW',0,None,None,0)
B_EBP = Beta('B_EBP',0,None,None,0)
B_EBC = Beta('B_EBC',0,None,None,0)
B_SL = Beta('B_SL',0,None,None,0)
B_ACC = Beta('B_ACC',0,None,None,0)
B_LUX = Beta('B_LUX',0,None,None,0)
B_PR = Beta('B_PR',0,None,None,0)
```

Definition of new variables. Method: adding columns to the database

Specification of the utility functions

```python
V1 = B_LDW * LDW_A + B_EBP * EBP_A + B_EBC * EBC_A + B_SL * SL_A + B_ACC * ACC_A + B_LUX * LUX_A + B_PR * PR_A
V2 = B_LDW * LDW_B + B_EBP * EBP_B + B_EBC * EBC_B + B_SL * SL_B + B_ACC * ACC_B + B_LUX * LUX_B + B_PR * PR_B
V3 = B_LDW * LDW_C + B_EBP * EBP_C + B_EBC * EBC_C + B_SL * SL_C + B_ACC * ACC_C + B_LUX * LUX_C + B_PR * PR_C
```

Associate the utility functions with the numbering of the alternatives

```python
V = {1: V1, 2: V2, 3: V3}
```

Associate the availability conditions with the alternatives

```python
AV1 = 1
AV2 = 1
AV3 = 1
```

```python
av = {1: AV1, 2: AV2, 3: AV3}
```

The contribution of the log likelihood function is the logarithm of a logit model

```python
logprob = bioLogLogit(V,av,CHOICE)
```

Biogeme

```python
biogeme  = bio.BIOGEME(database,logprob)
biogeme.modelName = "Attributes"
```

Running the estimation

```python
results = biogeme.estimate()
```

Read the results

```python
pandasResults = results.getEstimatedParameters()
pandasResults
```
